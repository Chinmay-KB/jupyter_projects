```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
tests=pd.read_csv('state_test.csv')
dict_values={}
tests.set_index(['states_tested_data/state','states_tested_data/updatedon'],inplace=True)
state_list=tests.index.get_level_values('states_tested_data/state').unique()
```

```python
total_test_list=[]
for states in state_list:
    total_test_list.append(tests.loc[states]['states_tested_data/totaltested'][-1])
    dict_values[states]=str(tests.loc[states]['states_tested_data/totaltested'][-1])
```

```python
plt.figure(figsize=(20, 3))
plt.bar(state_list,total_test_list)
plt.xticks(rotation=90)
plt.xlabel('States')
plt.ylabel('Total cumulative testing')
```
